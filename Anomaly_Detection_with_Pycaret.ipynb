```python
from pycaret.datasets import get_data
anomaly=get_data('anomaly')
anomaly.shape
```

```python
from pycaret.anomaly import *
exp_ano = setup(anomaly)
```

```python
#Create a model with k-nearest neighbours.
knn=create_model('knn')
```

```python
# Plot 3D model.
plot_model(knn)
```

Here, these yellow dots remain the anomaly activities and blues are the normal activity////.

```python
# Generate predictions using trained model:
knn_predictions= predict_model(knn, data =anomaly)
knn_predictions
```

```python
knn_results=assign_model(knn)
knn_results
```

```python
from pycaret.classification import *
Clf1 = setup(data = knn_results, target = 'Anomaly')
```

```python
# During the arrival of the model filter the values////:
best= compare_models(sort='Recall', )
```
